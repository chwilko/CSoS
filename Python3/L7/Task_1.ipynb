```python
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.insert(1, "..")

from packages.Levy_processes import *
```

```python
Image_width = 10
```

```python
Alphas = [0.25, 0.5, 1, 1.5]
Betas = [-0.5, 0, 0.5]
T = 1
M = 10
N = 1000
t = np.linspace(0, 1, N)

fig, axs = plt.subplots(nrows = len(Alphas), ncols = len(Betas), figsize = (2 * Image_width, 1.5 * Image_width))

for i, alpha in enumerate(Alphas):
    for j, beta in enumerate(Betas):
        X = Levy_processes(T, M, N, alpha, beta)
        ax = axs[i, j]
        ax.plot(t, X)
        ax.set_title(f"Alpha = {alpha}, Beta = {beta}")


```
