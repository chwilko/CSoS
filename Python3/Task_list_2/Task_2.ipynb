```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os

sys.path.insert(1, os.sep.join(["..", "packages"]))
from alphastable import multivariate_alphastable
```

```python
Image_width = 10
plt.rc('text', usetex = True) # latex ticks
plt.rc('text.latex')
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ['Computer Modern'],
})

EXTRA_SMALL = 10
SMALL_SIZE = 12
MEDIUM_SIZE = 15
BIGGER_SIZE = 20

plt.rc('font', size=EXTRA_SMALL)         # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

## Symmetric case

```python
N = 10 ** 4
gamma = [0.25, 0.125, 0.25, 0.25, 0.125, 0.25]
points = [[1, 0], [0.5, np.sqrt(3)/2],
            [-1/2, np.sqrt(3)/2], [-1, 0],
            [-1/2, -np.sqrt(3)/2], [1/2, -np.sqrt(3)/2]] 
alpha = 1.5
MA = multivariate_alphastable(alpha, gamma, points, N)
```

```python
k = 10
fig, ax = plt.subplots(figsize = (Image_width, Image_width))
ax.plot(MA[:, 0], MA[:, 1], '.');
ax.set_ylim([-k, k])
ax.set_xlim([-k, k])
ax.set_title("Symmetric case");
```

## Independent

```python
N = 10 ** 4
gamma = [0.1, 0.1, 0.1, 0.1]
points = [[np.cos(i), np.sin(i)] for i in np.arange(0, 2*np.pi, 2 * np.pi / len(gamma))]
alpha = 1.5
MA = multivariate_alphastable(alpha, gamma, points, N)
```

```python
k = 10
fig, ax = plt.subplots(figsize = (Image_width, Image_width))
ax.plot(MA[:, 0], MA[:, 1], '.');
ax.set_ylim([-k, k])
ax.set_xlim([-k, k])
ax.set_title("Independent case");
```

## Not symmetric with no independent components

```python
N = 10 ** 4
gamma = [0.1, 0.1, 0.1, 0.1]
angle_start = np.pi / 4
angle_stop = 2*np.pi
points = [[np.cos(i), np.sin(i)] for i in np.arange(angle_start, angle_stop, (angle_stop - angle_start) / len(gamma))]
alpha = 1.5
MA = multivariate_alphastable(alpha, gamma, points, N)
```

```python
k = 10
fig, ax = plt.subplots(figsize = (Image_width, Image_width))
ax.plot(MA[:, 0], MA[:, 1], '.');
ax.set_ylim([-k, k])
ax.set_xlim([-k, k])
ax.set_title("Not symmetric with no independent components");
```
